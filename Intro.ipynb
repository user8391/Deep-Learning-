```python
# SOFTMAX fUNCTION
import torch
import numpy as np
import torch.nn as nn # Torch lib for neural networks
import matplotlib.pyplot as plt

#Numpy (Manual Calculation for softmax)
z=[1,2,3]

ez=np.exp(z) # .exp function raises the input to the power of e
sum=np.dot(np.exp(z),[1,1,1]) # summing all the values
print(ez/sum) #softmax output
np.sum(ez/sum)
plt.yscale('log') # Logarthmic linear transformation of input to output
plt.scatter(z,ez/sum)
plt.show()
```

# **ENTROPY , CROSS** **ENTROPY**

```python
# Entropy - Uncertainty
import numpy as np
import matplotlib.pyplot as plt

# P of an event to happen = 0.25
p=[0.25,0.75] # 0.75 - Peobability of the event not to happen
entropy=0
for x in p:
  entropy-= x*np.log(x)
f"correct entropy {str(entropy)}"
```

```python
# Cross Entropy - Uncertainty (actual , predicted)
actual = [0.0,1.0]
predicted=[0.75,0.25]

cross=0
for p,q in actual,predicted:
  cross=cross-p*(np.log2(q))
"cross entropy "+str(cross)
```

```python
#binary Cross Entropy in Torch
import torch
import torch.nn.functional as F

# Inputs must be tensors
q_tensor = torch.tensor(predicted)
p_tensor = torch.tensor(actual)

# Binary Cross Entropy
F.binary_cross_entropy(q_tensor,p_tensor) #Sensitive to order of inputs
```

```python
# Min / Max & Argmin / Argmax
import numpy as np
import torch
import torch.nn as nn

# Numpy (vectors)
v=np.array([2,1,4,3,0.3,7])
#min/max
minval=np.min(v)
maxval=np.max(v)
print(minval,maxval,sep="\n")
#argmin/argmax
minpos=np.argmin(v)
maxpos=np.argmax(v)
print(minpos,maxpos,sep="\n")

# Numpy (matrices)
m=np.array([[1,2,3],[0,2,5]])

# For Matrix as a whole set of numbers
#min/max
minval=np.min(m)
minval_row=np.min(m,axis=0) # min value in each column
minval_col=np.min(m,axis=1) # min value in each row
maxval=np.max(m)
print(minval,maxval,minval_row,minval_col,sep="\n")
#argmin/argmax
minpos=np.argmin(m)
maxpos=np.argmax(m)
minval_row_pos=np.argmin(m,axis=0) # posetion of min value in each column
minval_col_pos=np.argmin(m,axis=1) # posetion of min value in each row
print(minpos,maxpos,minval_row_pos,minval_col_pos,sep="\n")

```

```python
# min ,max , argmin , argmax using Torch !!
# in vectors (1D Tensors) , works same as numpy , torch.min(v) , torch.max(v) and same for argmin and argmax

# in matrices (2D Tensors) - we do not need argmin for axis and all , the min
# returns a datatype that gives us both values and their indices (which indevidually are tensors)

matrix = torch.tensor([[2,3,1],[0,2,1]])

min1=torch.min(matrix) # 0D tensor
min2=torch.min(matrix,axis=0) # not a tensor

print(min1,min2,min2.values,sep="\n\n")


```

# T - test

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# parameters
n1 = 30 # samples of dataset 1
n2 = 40 # samples of dataset 2
# (mean offset)
mu1 = 1 # population mean of datastet 1
mu2 = 2 # population mean of datastet 2

# generating fake data
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

# plotting
plt.plot(np.zeros(n1),data1,'b.',label='data1')
plt.plot(np.ones(n2),data2,'r.',label='data2')
plt.legend()
plt.show()
```

```python
# t-test using stats package
# _ind -- Independent samples

t,p = stats.ttest_ind(data1,data2)
print(t,p)
```

```python
# Better way of visualizing
# to better visualize overlapping points .
plt.plot(0+np.random.randn(n1)/15,data1,'b.',label='data1')
plt.plot(1+np.random.randn(n2)/15,data2,'r.',label='data2')
plt.legend()
plt.show()
```

# Derivatives - sympy (very rich , but will not be used in a lot of DL , more specific to Derivatives & gradient discent)

```python
import sympy as sym
import sympy.plotting.plot as symplot # Symbolic math in python
```

```python
# Creating symbolic variables in sympy
x=sym.Symbol('x')

# Creating a function
fx = 2*x**2

# Computing its derivative
dfx=sym.diff(fx,x)
dfx
```

```python
# Visual representation
print(symplot(fx,(x,-1,1)))
print(symplot(dfx,(x,-1,1)))
```

Repeating with ReLU and Sigmoid

```python
# Create symbolic functions
relu = sym.Max(0,x)
sigmoid = 1/(1+sym.exp(-x)) # 1/(1+e^-x)
```

```python
# ReLU and Sigmoid
p = symplot(relu,(x,-4,4),show=False)
p.extend(symplot(sigmoid,(x,-4,4),show=False))
p.show()

# Their Derivatives
p = symplot(sym.diff(relu,x),(x,-4,4),show=False)
p.extend(symplot(sym.diff(sigmoid,x),(x,-4,4),show=False))
p.show()
```
